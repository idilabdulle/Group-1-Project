## Data exploration / cleaning for the Index of Multiple Deprivation dataset

This dataset is the Index of Multiple Deprivation, which is owned by the Ministry of Housing, Communities & Local Government.

Quote from dataset notes page:
About the English Indices of Deprivation 2025 (IoD25) 
The Indices of Deprivation 2025 provide a set of relative measures of deprivation for small areas (Lower-layer Super Output Areas) across England, based on seven domains of deprivation. The domains were combined using the following weights to produce the overall Index of Multiple Deprivation:
Income Deprivation (22.5%)
Employment Deprivation (22.5%)
Education, Skills and Training Deprivation (13.5%)
Health Deprivation and Disability (13.5%)
Crime (9.3%)
Barriers to Housing and Services (9.3%)
Living Environment Deprivation (9.3%)

Link to the webpage: https://www.gov.uk/government/statistics/english-indices-of-deprivation-2025
